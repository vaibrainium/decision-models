```python
means = list(np.arange(0,40,0.2))



prior = stats.norm(20, 4)

post = np.zeros(200)

for i, stimulus in enumerate(means):
    likelihood = stats.norm(stimulus, 5)
    
    
    
    post[i] = prior.pdf(stimulus)*likelihood.pdf(30)

post /= np.sum(post)
post /= 0.2
```
